```python
cols_order=["pq","graph","app","threads","malloc","time","cycles","work_cyc","galois","push","pop","emptypop","good","bad","empty"]
df=df[cols_order]
df=df[df.pq!="klsm1"]
def rename(x):
    if "klsm" in x:
        if "mq4" in x:
            curr=x.replace("klsm","")
            curr=curr.replace("mq4","")
            k=int(curr)
            return 'klsm{:04d}mq4'.format(k)
        if "mq" in x:
            curr=x.replace("klsm","")
            curr=curr.replace("mq","")
            k=int(curr)
            return 'klsm{:04d}mq'.format(k)
        if "capq" in x:
            curr=x.replace("klsm","")
            curr=curr.replace("capq","")
            k=int(curr)
            return 'klsm{:04d}capq'.format(k)
        else:
            curr=x.replace("klsm","")
            k=int(curr)
            return 'klsm{:04d}'.format(k)
    else:
        return x
    
df["pq"]=df["pq"].apply(rename)

```

```python
colors={'klsm0256':'violet',
        'klsm0512':'darkviolet',
        'klsm1024':'rebeccapurple',
        'klsm2048':'slateblue',
        'klsm4096':'darkblue',
        'multiqueue1':'brown',
        'multiqueue4':'darkred',
        'obim':'lightslategrey',
        'capq':'orange',
        'klsm0256mq':'teal',
        'klsm0256mq4':'lightseagreen',
        'klsm0064mq':'mediumseagreen',
        'klsm0064mq4':'springgreen',
        'klsm4096mq':'forestgreen',
        'klsm4096mq4':'limegreen',
        'spraylist':'pink',
        0:'orange',
        1:'teal',
        2:'k',
        3:'c',
        4:'y'
       }
```

```python
targets=["work_cyc","push","pop","emptypop","galois"]

def plot_stacked_bar(data,graph,threads):
    
    data=data[data.graph==graph]
    data=data[data.threads==threads]
    #display(data)
    names=data["pq"].unique()
    plt.figure(figsize=(15,6))
    for index,pq in enumerate(data["pq"].unique()):
        pq_data=data[data.pq==pq]
        #display(pq_data)
        bottom=0
        for color,target in enumerate(targets):
            targetMean=pq_data[target].mean()
            targetStd=pq_data[target].std()
            plt.bar(index, targetMean, 0.35,color=colors[color], edgecolor='white',bottom=bottom, yerr=targetStd)
            bottom+=targetMean
        plt.legend(["work_cyc","push","pop","emptypop","galois"])
    plt.xticks(range(len(names)), names,rotation=45)
    plt.xlabel("Queue")
    plt.ylabel("Cycles")
    #plt.yscale('log')
    plt.title("Total Work plot of graph \""+graph+"\" for "+str(threads)+" threads")
    plt.show()
```

```python
def plot_good_bad_bar(data,graph,threads):
    
    data=data[data.graph==graph]
    data=data[data.threads==threads]
    #display(data)
    names=data["pq"].unique()
    plt.figure(figsize=(15,6))
    curr=data.groupby(['pq'])
    goodMean=curr["good"].mean()
    badMean=curr["bad"].mean()
    emptyMean=curr["empty"].mean()
    goodStd=curr["good"].std()
    badStd=curr["bad"].std()
    emptyStd=curr["empty"].std()
    p1 = plt.bar(goodMean.index, goodMean, 0.35,color=colors[0], edgecolor='white', yerr=goodStd,label='good')
    p2 = plt.bar( badMean.index,badMean, 0.35, color=colors[1], edgecolor='white',bottom=goodMean, yerr=badStd,label="bad")
    p3 = plt.bar( emptyMean.index,emptyMean, 0.35, color=colors[2], edgecolor='white',bottom=goodMean+badMean, yerr=emptyStd,label="empty")
    plt.legend((p1[0], p2[0],p3[0]), ('good', 'bad','empty'))
    plt.xticks(range(len(names)), names,rotation=45)
    plt.xlabel("Queue")
    plt.ylabel("Cycles")
    plt.title("Good/Bad/Empty plot of graph \""+graph+"\" for "+str(threads)+" threads")
    plt.show()
```

```python

def plot_graph(data,title):
    plt.figure(figsize=(15,10))
    for pq in data.pq.unique():
        curr=data[data.pq==pq]
        for malloc in curr.malloc.unique():
            curr2=curr[curr.malloc==malloc]
            #display(curr)
            currMean=curr2.groupby(['threads']).mean()
            #display(currMean)
            currStd=curr2.groupby(['threads']).std()
            #display(curr.groupby(['threads']).groups.keys())
            currThreads=curr2.groupby(['threads']).groups.keys()
            plt.errorbar(currThreads, currMean["time"],yerr=currStd["time"],label=pq,color=colors[pq])#+" "+malloc
    plt.title(title)
    plt.yscale('log')
    plt.ylabel('Time')
    plt.xscale('log')
    plt.xlabel('Threads')
    plt.legend()
    plt.grid()
    plt.show()
def plot_graphs(data):
    for graph in data["graph"].unique():
        plot_graph(data[data.graph==graph],'SSSP, Time vs. Threads on graph: "'+graph+'"')
        plot_stacked_bar(df,graph,88)
        plot_good_bad_bar(df,graph,88)
plot_graphs(df)
#plot_graph(df[df.graph=="clique4000"],"clique4000")

```
